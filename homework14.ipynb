```python
def ejection(df, param):
    q1 = df[param].quantile(0.25)
    q3 = df[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = df[df[param].between(lower_bound, upper_bound, inclusive=True)]
    df_result = pd.concat([df, remove_outliers]).drop_duplicates(keep=False)[[param]]
    df_result = df_result.dropna()
    return df_result[param].tolist()
```

```python
outliers_rec_temp = ejection(df, 'rectal temperature')
outliers_rec_temp
```

```python
outliers_pulse = ejection(df, 'pulse')
outliers_pulse
```

```python
outliers_resp_rate = ejection(df, 'respiratory rate')
outliers_resp_rate 
```

```python
df[df['pulse'].isin(outliers_pulse)]
```
